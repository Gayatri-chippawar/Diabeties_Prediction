**Importing the Libraries**

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.linear_model import LogisticRegression
```

**Loading the Dataset**

```python
data = pd.read_csv('../datasets/diabetes.xls')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.shape
```

**Dividing the Dataset**

```python
x = data.drop(columns='Outcome',axis=1)
y = data['Outcome']
```

```python
print(x)
```

```python
print(y)
```

**Data Standardization**

```python
print(x.std().sum())
```

```python
scaler = StandardScaler()
```

```python
x_std = scaler.fit_transform(x)
```

```python
print(x_std.std().sum())
```

**Splitting The dataset**

```python
x_train,x_test,y_train,y_test = train_test_split(x_std,y,test_size=0.2,stratify=y,random_state=42)
```

```python
print(x_std.shape,x_train.shape,x_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

**Loading the model**

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```

```python
training_prediction = model.predict(x_train)
training_acc = accuracy_score(y_train,training_prediction)
```

```python
print(training_acc)
```

```python
testing_prediction = model.predict(x_test)
testing_acc = accuracy_score(y_test,testing_prediction)
print(testing_acc)
```

**Cross Validation**

```python
x_train2,x_test2,y_train2,y_test2 = train_test_split(x_std,y,test_size=0.2,stratify=y,random_state=45)
```

```python
model.fit(x_train2,y_train2)
```

```python
training_prediction2 = model.predict(x_train2)
training_acc2 = accuracy_score(y_train2,training_prediction2)
print(training_acc2)
```

```python
testing_prediction2 = model.predict(x_test2)
testing_acc2 = accuracy_score(y_test2,testing_prediction2)
print(testing_acc2)
```

```python
x_train3,x_test3,y_train3,y_test3 = train_test_split(x_std,y,test_size=0.2,stratify=y,random_state=55)
```

```python
model.fit(x_train3,y_train3)
```

```python
training_prediction3 = model.predict(x_train3)
training_acc3 = accuracy_score(y_train3,training_prediction3)
print(training_acc3)
```

```python
testing_prediction3 = model.predict(x_test3)
testing_acc3 = accuracy_score(y_test3,testing_prediction3)
print(testing_acc3)
```

```python
print("Avg training accuracy of svm :",(training_acc+training_acc2+training_acc3)/3)
```

```python
print("Avg testing accuracy of svm :",(testing_acc+testing_acc2+testing_acc3)/3)
```
